#### Loading Ratings Dataset

```python
# # uploading in dataset
# from google.colab import files
# uploaded = file.upload()
```

```python
# Reading the dataset
ratings = pd.read_csv('ratings.csv')
```

```python
# first few rows
ratings.head()
```

```python
# dataset info
ratings.info()
```

#### Merging movies and ratings

```python
# merging datasets
ratings = pd.merge(movies, ratings).drop(['genres', 'timestamp'], axis=1)
```

```python
# first few rows
ratings.head()
```

```python
# df info
ratings.info()
```

#### New Dataframe of UserID by Movies

```python
# creating userId by Movies dataframe
ratings = ratings.pivot_table(index = ['userId'], columns = ['title'], values = 'rating')
```

```python
# first few rows
ratings.head()
```

```python
# Dataframe dimension
ratings.shape
```

```python
# dropping columns with few ratings
ratings = ratings.dropna(thresh = 10, axis=1)
```

```python
# Dataframe dimension
ratings.shape
```

```python
# first few rows
ratings.head()
```

#### Item Based Collaborative Filtering

```python
# correlation matrix
corrMatrix = ratings.corr()
corrMatrix.head()
```

```python
# find similar movies
corrMatrix['127 Hours (2010)'].sort_values(ascending = False)
```

#### User Based Collaborative Filtering

```python
# transpose
ratings.T
```

```python
# correlation matrix for users
corrMatrix = ratings.T.corr()
corrMatrix.head()
```

```python
# Finding similar user
corrMatrix.iloc[:,1].sort_values(ascending = False)
```
